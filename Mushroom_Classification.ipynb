```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Rupeshbhardwaj002/Mushroom_Classification_Datstet/refs/heads/main/mushrooms.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
model = LabelEncoder()
l = df.columns.size
for i in range(0 , l):
  df[df.columns[i]] = model.fit_transform(df[df.columns[i]])
```

```python
df.head()
```

```python
df.info()
```

```python
x = df[['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']]
y = df['class']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(max_iter=200)
lg.fit(x_train, y_train)
y_predict = lg.predict(x_test)
from sklearn.metrics import accuracy_score, confusion_matrix

accuarcy = accuracy_score(y_predict, y_test)

confusion = confusion_matrix(y_predict, y_test)
print("Accuracy", accuarcy*100)
print("Confusion Matrix", confusion)
```

```python
y_predict = lg.predict(x_train)
from sklearn.metrics import accuracy_score, confusion_matrix

accuarcy = accuracy_score(y_predict, y_train)

confusion = confusion_matrix(y_predict, y_train)
print("Accuracy", accuarcy*100)
print("Confusion Matrix", confusion)
```

```python
from sklearn.svm import SVC
sv = SVC(kernel='linear')
sv.fit(x_train, y_train)
y_predict = sv.predict(x_test)
from sklearn.metrics import accuracy_score, confusion_matrix

accuarcy = accuracy_score(y_predict, y_test)

confusion = confusion_matrix(y_predict, y_test)
print("Accuracy", accuarcy*100)
print("Confusion Matrix", confusion)
```
